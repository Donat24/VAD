```python
import torch
import torchaudio
import matplotlib.pyplot as plt
from IPython.display import Audio
```

```python
path=r"C:\Users\Jonas\Documents\Masterarbeit\Scripte\Samples_Processed\Breakbeat 01.wav"
tensor, sr = torchaudio.load(path)

#print
tensor, sr
```

```python
#PLOT

#x-Achse
time_axis = torch.arange(0,tensor.shape[-1]) / sr
time_axis.numpy()

#y-Achse
mono_chanell = 0
y = tensor[mono_chanell].numpy()

plt.plot(time_axis,y)
plt.xlabel("Zeit in Sekunden")
plt.ylabel("Amplitute")
plt.title("Waveform Plot")
```

```python
#y-Achse
mono_chanell = 0
values = tensor[mono_chanell].numpy()

plt.specgram(values,Fs=sr)
plt.show()
```
